## Usando células de código

A linguagem padrão dos Jupyter Notebooks é Python. Podemos inserir facilmente células com código Python para serem interpretadas junto com o restante do notebook.

Porém, muitas outras linguagens de programação podem ser utilizadas bastando para isso instalar o Kernel da linguagem desejada.

```python
mensagem = "Hello Jupyter!"
print(mensagem)
```

Uma vantagem das células de código nos Jupyter Notebooks é que o resultado da última linha de uma célula é exibido automaticamente no notebook sem precisar usar a função `print()`.

Exemplo:

```python
nome = "Gustavo"
"Olá, " + nome
```

Isso é muito bom para cálculos onde geralmente a última linha é a que contém o resultado.

Exemplo:

```python
nota1 = 5.5
nota2 = 6.0
nota3 = 8.0
somatorio = nota1 + nota2 + nota3

# média das 3 notas
somatorio / 3
```

### Compartilhamento de memória entre células

As células de código em um Jupyter Notebook não são "uma ilha"! Elas compartilham o mesmo ambiente de execução - elas fazem parte de um mesmo programa e compartilham memória.

As variáveis criadas na célula anterior podem ser acessadas em células posteriores. Elas são compartilhadas dentro do notebook.

Exemplo:

```python
# note que nesta célula não foi criada a variável somatorio
# somatorio foi criada em uma célula anterior a esta
# mas mesmo assim está na memória, disponível para a célula atual usar

print(f"O somatório das notas é {somatorio}")
print(f"A média é: {somatorio / 3}")
```
