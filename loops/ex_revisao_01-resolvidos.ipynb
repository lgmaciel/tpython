1. Escreva um programa que
    - receba um número `x` (inteiro) pelo teclado
    - conte de 1 até `x`, imprimindo na tela cada um dos números contados

````python
x = input("Digite um número inteiro: ")
for i in range(1, int(x)+1):
    print(i)
````

3. Escreva um programa parecido com o anterior, mas que gere uma lista no seguinte formato:

```
Lista de compras
----------------
1. água
2. manteiga
3. açúcar
4. pão
...
```

````python
numero = 1

print("Lista de Compras")
print("----------------")

while True:
    lista_de_compras = input("Digite as compras: ")

    if lista_de_compras == "end":
        break
    print(str(numero) + ". " + lista_de_compras)
    numero = numero + 1

````

````python
numero = 1
lista_de_compras = "" # inicializando para que entre no loop while na primeira iteração

print("Lista de Compras")
print("----------------")

while lista_de_compras != "end":
    lista_de_compras = input("Digite as compras: ")
    print(str(numero) + ". " + lista_de_compras)
    numero = numero + 1

````

````python
compras = ['água', 'manteiga', 'açúcar', 'pão', 'refri']

print("Lista de Compras")
print("----------------")

for i in range(len(compras)):
    print(f"{i+1}. {compras[i]}")
````
