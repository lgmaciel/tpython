## Exercício 1: Contagem Simples
Escreva um programa que imprima os números de 1 a 10. Este exercício introduz o conceito básico de um loop `for` para contagem.



````python
for i in range(1, 11):
    print(i)
````

E se não estivermos interessados nos números em si, mas apenas na contagem. Por exemplo, para executar uma mesma operação `n` vezes?

````python
# Quero imprimir "boa noite" 10 vezes
for _ in range(10):
    print("Boa noite")
````

## Exercício 4: Soma de Números
Escreva um programa que calcule a soma dos números de 1 a 100. Aqui, você aprenderá a usar uma variável acumuladora.


````python
soma = 0
for i in range(1, 101):
    soma += i
print("A soma é:", soma)
````

Exemplo: Total de uma lista de compras

````python
valor_dos_itens = [5.5, 20.0, 10.99, 1.99, 8.0]

soma = 0
for i in valor_dos_itens:
    soma += i
print(f"A soma é: {soma:.2f}", )
````

Exemplo: Quero contar quantas vezes aparece 'João' numa lista.

````python
nomes = ["João", "Ana", "João", "João", "String","Float","João","André", "Ricardo"]

quantos = 0 # variável contadora
for nome in nomes:
    if nome == "João":
        quantos += 1

print(quantos)
````

Imprimir a quantidade de valores em branco ""

````python
nomes = ["João", "NÃO SEI", "João", "João", "NÃO SEI","André", "João","André", "NÃO SEI"]

quantos = 0 # variável contadora
for nome in nomes:
    if nome == "André":
        quantos += 1

print(quantos)
````

passo a passo

````python
nomes = ["João", "NÃO SEI", "João", "João", "NÃO SEI","André", "João","André", "NÃO SEI"]

contagem = 0
for nome in nomes:
    if nome == "NÃO SEI":
        contagem += 1

print(contagem)
````

## Exercício 5: Contagem Regressiva
Implemente um programa que imprima uma contagem regressiva de 10 a 1. Este exercício reforça o uso do `range` com valores decrescentes.


````python
for i in range(10, 0, -1):
    print(i)
````

Coisas pra fazer de trás pra frente...

````python
# Imprimir os valores de uma lista de trás pra frente
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza"]
for i in range(len(nomes), 0, -1):
    print(nomes[i-1])
````

alternativa sem usar variável contadora

````python
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza","Fernanda"]
for i in range(len(nomes)):    
    print(nomes.pop())
````

## Exercício 6: Multiplicação de Elementos de uma Lista
Dada uma lista de números `[2, 4, 6, 8]`, escreva um programa que imprima o dobro de cada número. Este exercício combina iteração sobre listas e operações matemáticas.


````python
numeros = [2, 4, 6, 8]
for numero in numeros:
    print(numero * 2)
````

Trocando o operador por uma função. Como aplicar uma função a uma série de valores?

````python
valores = [2, 4, 6, 8, 10, 12, 14]

def y(x):
    return 30 * x + 4 # ax + b

for x in valores:    
    print(y(x))

print("\nde novo...")

for i in range(len(valores)):
    val_y = y(valores[i])
    print(val_y)
````

Mesma ideia: Aplicar uma função a uma lista de valores.

````python
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza","Fernanda"]

def gerar_email(nome_usuario):
    return f"{nome_usuario.lower()}@mtm.ia.br"

for nome in nomes:
    print(gerar_email(nome))
````

Agora nomes e apelidos em listas diferentes.

````python
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza","Fernanda"]
apelidos = ["Aninha", "Bia", "Carol", "Didi", "Iza", "Fê"]

for i in range(len(nomes)):
    print(f"{nomes[i]} - {apelidos[i]}")
````

Nomes e idades

````python
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza","Fernanda"]
idades = [20, 30, 21, 34, 45, 60]

for i in range(len(nomes)):
    print(f"{nomes[i]}, Idade: {idades[i]}, {gerar_email(nomes[i])}")
````

````python
nomes = ["Ana","Beatriz","Carolina","Diana","Eliza","Fernanda"]
apelidos = ["Aninha", "Bia", "Carol", "Didi", "Iza", "Fê"]

for i in range(len(nomes)):
    print(f"{nomes[i]} - Idade: {len(apelidos[i])}")
````

## Exercício 7: Contagem de Caracteres em Strings
Peça ao usuário para digitar uma palavra e use um loop `for` para contar quantas letras "a" existem na palavra. Este exercício introduz a iteração sobre strings.

```python
palavra = input("Digite uma palavra: ")
contador = 0
for letra in palavra:
    if letra == "a":
        contador += 1
print("A palavra contém", contador, "letras 'a'.")
```
