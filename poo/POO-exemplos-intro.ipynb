Imagine que a tarefa é modelar um "disco-voador".

```python
disco_voador = "diâmetro: 20, nome: ISSXXZ, turbinas: formato: cilindrico: quantidade:9"
```

```python
pos_nome = disco_voador.find("nome:")
disco_voador[pos_nome+6:]
```

O melhor é modelar usando várias variáveis

```python
disco_voador_nome = "ISSXXZ"
disco_voador_diametro = 20

disco_voador_turbinas_formato = "cilindrico"
disco_voador_turbinas_quantidade = 9
disco_voador_turbinas_funcionando = False

disco_voador_voando = False
disco_voador_velocidade = 0

def mostra_nome_disco_voador():
    print(disco_voador_nome)

def mostra_velocidade_disco_voador():
    print(disco_voador_velocidade)

def disco_voador_decolar():
    global disco_voador_turbinas_funcionando, disco_voador_velocidade, disco_voador_voando
    disco_voador_turbinas_funcionando = True
    disco_voador_velocidade = 10
    disco_voador_voando = True

def disco_voador_aterrizar():
    global disco_voador_turbinas_funcionando, disco_voador_velocidade, disco_voador_voando
    disco_voador_turbinas_funcionando = False
    disco_voador_velocidade = 0
    disco_voador_voando = False

```

```python
mostra_nome_disco_voador()
mostra_velocidade_disco_voador()
disco_voador_decolar()

mostra_nome_disco_voador()
mostra_velocidade_disco_voador()
```

Imagine que eu queira fazer o disco-voador aterrizar... Mas *sem usar as funções*

```python
disco_voador_turbinas_funcionando = False
disco_voador_velocidade = 0
disco_voador_voando = False
```

Mas... pense que poderíamos também fazer isso *errado*

Vamos agora decolar...

```python
disco_voador_turbinas_funcionando = False
disco_voador_voando = True
```

assim eu "estraguei" o modelo. Ele perdeu sua consistência.

Por isso, usar as funções garantirá a coerência interna do "objeto" (ele ainda não é um objeto, ok?)

É mais correto (e seguro/consistente) decolar assim:

```python
disco_voador_decolar()
```

## Criando objetos

Começamos definindo a *Classe* do objeto.

> *Classe*, aqui, é sinônimo de *Conjunto*. Isso porque estamos definindo características de quaisquer objetos e não de um objeto em particular (um *exemplar* da Classe - ou, *instância* da Classe)

```python
class Disco_Voador:
    # Propriedades/Estados
    nome = ""
    diametro = 0
    turbinas_formato = ""
    turbinas_quantidade = 0
    turbinas_funcionando = False
    voando = False
    velocidade = 0
    
```

#### Criando Objetos (*Instâncias da Classe*)

```python
disco1 = Disco_Voador()
disco2 = Disco_Voador()
```

Acessando propriedades...

```python
disco1.nome = "XXZZA"
disco2.nome = "YZYZB"
print(disco1.nome)
print(disco2.nome)
```

### Adicionando métodos

```python
class Disco_Voador:
    # Propriedades/Estados
    nome = ""
    diametro = 0
    turbinas_formato = ""
    turbinas_quantidade = 0
    turbinas_funcionando = False
    voando = False
    velocidade = 0
    
    def mostra_nome(self):
        print(self.nome)
    
    def mostra_velocidade(self):
        print(self.velocidade)

    def decolar(self):
        self.turbinas_funcionando = True
        self.velocidade = 10
        self.voando = True
    
    def aterrizar(self):
        self.turbinas_funcionando = False
        self.velocidade = 0
        self.voando = False
```

```python
disco1 = Disco_Voador()
disco2 = Disco_Voador()

disco1.nome = "XXZZA"
disco2.nome = "YZYZB"

disco1.mostra_nome()
disco1.mostra_velocidade()

disco2.mostra_nome()
disco2.mostra_velocidade()

disco1.decolar()
disco1.mostra_nome()
disco1.mostra_velocidade()
print(disco1.turbinas_funcionando)

disco2.mostra_nome()
disco2.mostra_velocidade()
```
