## Conta-corrente

```python
class Conta_Corrente:
    def __init__(self, numero, titular):
        self.saldo = float()
        self.numero = numero
        self.titular = titular
    
    def __str__(self):
        return f"cc: {self.numero} ({self.saldo}) titular: {self.titular}"

    def depositar(self, valor):
        self.saldo += valor

    def sacar(self, valor):
        if self.saldo >= valor: 
            self.saldo -= valor
            return True # conseguiu sacar
        # não conseguiu sacar
        return False
```

```python
cc1 = Conta_Corrente(123893, "L G Maciel")
cc2 = Conta_Corrente(123432, "Cesar Ody")

print(cc1)
print(cc2)

# Movimentação
cc1.depositar(200.00)
cc2.depositar(120.00)

if not cc1.sacar(50.00):
    print("Saldo insuficiente")

if not cc1.sacar(200.00):
    print("Saldo insuficiente")

print("Saldos das contas:")
print(cc1)
print(cc2)
```
