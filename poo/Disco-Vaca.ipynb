# Disco Vaca

O que vacas tem? Quais suas propriedades? Seus estados?

Pergunta-chave: e de tudo o que vaca tem, o que importa pra o nosso problema?

```python
# todas as nossas Vacas terão nome
class Vaca:
    nome = ""
```

Testando. Vamos instanciar (criar) duas vacas

```python
# Aqui estamos criando 2 objetos do tipo Vaca
vaca1 = Vaca()
vaca2 = Vaca()
```

Agora, queremos poder dar nome para uma Vaca.

Para isso, precisamos criar o método que muda o nome. Ele vai receber um nome e atribuir esse nome à vaca.

```python
# todas as nossas Vacas terão nome
class Vaca:
    nome = ""

    def nome(self, novo_nome):
        self.nome = novo_nome
```

usando o método `Vaca.nome(novo_nome)`

```python
vaca1 = Vaca()
vaca1.nome("Mimosa")
vaca2 = Vaca()
vaca2.nome("Malhada")
```

Agora vamos acrescentar um método para "ler"/"acessar" o valor da propriedade "nome".


```python
# todas as nossas Vacas terão nome
class Vaca:
    nome = ""

    def set_nome(self, novo_nome):
        self.nome = novo_nome

    def get_nome(self):
        return self.nome
```

Como usar esse método que 'acessa' uma propriedade.

```python
vaca1 = Vaca()
vaca1.set_nome("Mimosa")
vaca2 = Vaca()
vaca2.set_nome("Malhada")

vaca1.get_nome()
```

- **getters**: Métodos que acessam propriedaes de um objeto.
- **setters**: Métodos que definem ("setam") propriedades


Agora, acrescentando uma propriedade *cor*

```python
# todas as nossas Vacas terão nome
class Vaca:
    nome = ""
    cor = ""

    # Definir getters e setters
    def set_nome(self, novo_nome):
        self.nome = novo_nome

    def get_nome(self):
        return self.nome

    def set_cor(self, nova_cor):
        self.cor = nova_cor
    
    def get_cor(self):
        return self.cor

    
```

Usando os novos métodos...

```python
vaca1 = Vaca()
vaca1.set_nome("Mimosa")
vaca1.set_cor("Preta")

vaca2 = Vaca()
vaca2.set_nome("Malhada")
vaca2.set_cor("Marrom/Branco")

print("Vacas que estão no pasto:")

print(vaca1.get_nome())
print(vaca1.get_cor())

print(vaca2.get_nome())
print(vaca2.get_cor())
```

Agora a vaca vai mugir...

```python
class Vaca:
    nome = ""
    cor = ""

    # Definir getters e setters
    def set_nome(self, novo_nome):
        self.nome = novo_nome

    def get_nome(self):
        return self.nome

    def set_cor(self, nova_cor):
        self.cor = nova_cor
    
    def get_cor(self):
        return self.cor

    def mugir(self):
        print("Muuuuu!!!")
```

Fazendo a vaca mugir:

```python
# Criando duas novas vacas...
vaca1 = Vaca()
vaca2 = Vaca()

# Fazendo uma delas mugir
vaca1.mugir()
```

Fazendo a vaca pastar, mas só quando ela tem fome.

Para isso, vamos acrescentar um *estado* à Vaca: **temFome**

```python
class Vaca:
    nome = ""
    cor = ""
    tem_fome = True # vacas "nascem" com fome
    pastando = False # vacas não nascem comendo

    # Definir getters e setters
    def set_nome(self, novo_nome):
        self.nome = novo_nome

    def get_nome(self):
        return self.nome

    def set_cor(self, nova_cor):
        self.cor = nova_cor
    
    def get_cor(self):
        return self.cor

    def fique_com_fome(self):
        self.tem_fome = True

    def fique_satisfeita(self):
        self.tem_fome = False

    def mugir(self):
        print("Muuuuu!!!")
    
    def pastar(self):
        if self.tem_fome:
            self.pastando = True
            print(f"{self.nome} pastando...")
            self.fique_satisfeita()
        else:
            print("Já estou satisfeita.")
```

```python
vaca1 = Vaca()
vaca2 = Vaca()

vaca1.set_nome("Mimosa")
vaca2.set_nome("Bondosa")

vaca1.pastar()
vaca1.pastar()
vaca2.pastar()
```

## Disco-voador capturando Vacas

Vamos criar uma classe Disco_Voador e fazer com que objetos dessa classe capturem (abduzam) vacas.

A idéia vai ser simples. Discos-Voadores terão uma `list()` como propriedade principal e essa `list()` é que vai conter as vacas. 

```python
class Vaca:
    pass

class Disco_Voador:
    pass

disco = Disco_Voador()
v1 = Vaca()
v2 = Vaca()

print(disco)
print(v1)
print(v2)
```

Note que, mesmo usando classes "vazias" já é o suficiente para definir um **tipo** para os objetos.

```python
print(type(v1))
print(type(disco))
```

Agora vamos acrescentar *propriedades de instância* às classes. Isso quer dizer que vamos definir nas classes quais propriedades cada instância (objeto) criada a partir delas terão.

```python
class Vaca:
    nome = str()

class Disco_Voador:
    sala_branca = list()

disco = Disco_Voador()
v1 = Vaca()
v2 = Vaca()

# Usando as propriedades de instância
v1.nome = "Mimosa"
v2.nome = "Malhada"

# Exibindo as propriedades...
print(v1.nome)
print(v2.nome)
print(f"Quant. vacas na sala branca: {len(disco.sala_branca)}")
```

#### Melhorando a impressão dos objetos.

Vamos definir um método especial chamado `__str__` que diz ao Python como converter em string um objeto. Esse método é usado pela função `print()`, por exemplo, e vai modificar o modo como os objetos são exibidos na tela.

```python
class Vaca:
    nome = str()

    def __str__(self):
        return f"Sou a {self.nome}"

class Disco_Voador:
    sala_branca = list()

disco = Disco_Voador()
v1 = Vaca()
v2 = Vaca()

# Usando as propriedades de instância
v1.nome = "Mimosa"
v2.nome = "Malhada"

print(v1)
print(v2)
```

### Abduzir vacas (finalmente...)

```python
class Vaca:
    nome = str()

    def __str__(self):
        return f"Sou a {self.nome}"

class Disco_Voador:
    sala_branca = list()

    def abduzir(self, uma_vaca):
        self.sala_branca.append(uma_vaca)

disco = Disco_Voador()
uma_vaca = Vaca()
uma_vaca.nome = "Mimosa"

print(len(disco.sala_branca)) # sala branca deve estar vazia

disco.abduzir(uma_vaca) # vaca vai para a sala branca

print(len(disco.sala_branca)) # sala branca deve conter 1 vaca
```

### Melhorando o **Encapsulamento** dos objetos

```python
class Vaca:

    # Método construtor
    def __init__(self):
        self.nome = str()

    def set_nome(self, nome):
        if type(nome) == str and len(nome) < 10:
            self.nome = nome

    def __str__(self):
        return f"Sou a {self.nome}"

class Disco_Voador:
    
    # Método construtor
    def __init__(self):
        self.sala_branca = list()

    def lotacao_sala_branca(self):
        return len(self.sala_branca)
    
    def abduzir(self, uma_vaca):
        self.sala_branca.append(uma_vaca)
```

```python
disco = Disco_Voador()
uma_vaca = Vaca()
uma_vaca.set_nome("Mimosinha")

print(disco.lotacao_sala_branca()) # sala branca deve estar vazia
disco.abduzir(uma_vaca) # vaca vai para a sala branca
print(disco.lotacao_sala_branca()) # sala branca deve conter 1 vaca
```
