```python
vidas = 3
pontos = 0
```

```python
def mostra_status():
    
    print("Status")
    if vidas > 0:
        print(vidas * "❤️")
    else:
        print("☠️")

    print(f"💰 {pontos}")
```

```python
# Perder uma vida
def perder_vida():
    global vidas
    if vidas > 0:
        vidas = vidas - 1
```

```python
# Ganhar uma vida
def ganhar_vida():
    global vidas
    vidas = vidas + 1
```

```python
def ganhar_pontos():
    global pontos
    pontos = pontos + 10

def perder_pontos():
    global pontos
    pontos = pontos - 10
```

```python
mostra_status()
```

```python
def senha_valida(senha):
    return len(senha) >= 8
```

```python
print("Você chegou na porta do Dungeon")
print("Acerte a senha para entrar")

senha_digitada = input("Qual é a senha? ")
if senha_valida(senha_digitada):
    ganhar_pontos()
    mostra_status()
    print("Parabéns! Pode entrar no castelo.")
else:
    perder_vida()
    mostra_status()
    print("Errado. Você foi punido. Menos uma vida.")
```
