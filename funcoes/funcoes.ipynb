# Funções

```python
# O dobro de algum valor numérico
valor = 10          # entrada 
dobro = valor * 2   # procesamento 
print(dobro)        # saída
```

Definição da função

```python
def dobro():
    valor = 10          # entrada 
    dobro = valor * 2   # procesamento 
    print(dobro)        # saída
```

Chamada da função

```python
dobro()
```

Múltiplas chamdas da função...

```python
dobro()
dobro()
print("mais uma vez...")
dobro()
```

Generalizando a função usando parâmetros

```python
def dobro(valor):       # entrada               
    dobro = valor * 2   # procesamento 
    print(dobro)        # saída
```

Chamando a função passando valores

```python
dobro(0)
dobro(1)
dobro(2)
dobro(3)
dobro(4)
```

Retornando um valor

```python
def dobro(valor):       # entrada               
    dobro = valor * 2   # procesamento 
    return dobro        # saída
```

Usando o valor de retorno de uma função

```python
resposta = dobro(2)
```

```python
print(resposta)
```

Exemplo prático: Validação de senha

```python
def validar_senha(senha):
    contagem = len(senha)
    if contagem >= 8:
        print("Senha válida")
    else:
        print("Senha inválida")
```

Uso:

```python
validar_senha("abcd")
validar_senha("12345678")
validar_senha("abracadabra")
validar_senha("12345")
```

```python
senha = "abcde"
contagem = len(senha)
if contagem >= 8:
    print("Senha válida")
else:
    print("Senha inválida")

senha = "abcdeabcde"
contagem = len(senha)
if contagem >= 8:
    print("Senha válida")
else:
    print("Senha inválida")

senha = "123456"
contagem = len(senha)
if contagem >= 8:
    print("Senha válida")
else:
    print("Senha inválida")

senha = "abcde3123131"
contagem = len(senha)
if contagem >= 8:
    print("Senha válida")
else:
    print("Senha inválida")
```

Outra versão da função `validar_senha()`

```python
def senha_valida(senha):
    return len(senha) >= 8
```

```python
v1 = senha_valida("abcd")
v2 = senha_valida("12345678")
v3 = senha_valida("abracadabra")
v4 = senha_valida("12345")

print(v1, v2, v3, v4)
```

Uso com `if/else`

```python
senha_digitada = input("Digite sua senha: ")

if senha_valida(senha_digitada):
    print("Ok. Pode continuar")
else:
    print("Sua senha é muito fraca.")

print("Fim do programa")
```
