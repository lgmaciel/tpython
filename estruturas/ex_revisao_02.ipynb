# Exercícios com listas

O programa abaixo gera uma lista de 1000 valores aleatórios entre 1 e 1000, como mostra o segundo `print()`.

Tarefa:

Complete o programa para que ele mostre:

1. a soma de todos os números da lista gerada
1. a média aritmética dos valores da lista
1. quantos múltiplos de 2 a lista gerada contém
1. quantos múltiplos de 20 a lista gerada contém

> Como a lista contém valores diferentes a cada vez que o programa roda, os resultados também serão diferentes a cada vez.

Você pode responder tudo em um mesmo programa ou copiar a célula abaixo várias vezes e responder cada ítem individualmente. Faça como preferir.

```python
import random
numeros = []
for i in range(10000):
    numeros.append(random.randrange(1,1001))

print("Lista Gerada")
print(numeros)
print()
print(f"Tamanho da lista: {len(numeros)}")

# Complete o programa aqui
# Você vai usar a lista numeros para responder as questões
```
